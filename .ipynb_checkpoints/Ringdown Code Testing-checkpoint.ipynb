# Example code for running ringdown fits.

## Important takeaways:
- Seeding the fit with a good approximate frequency helps a lot. Multiple frequency fit iterations can help with this.
- For noisy, low SNR data, multiple ringdown timestreams are necessary to get a good estimate of Q.
- Logfit is faster, but exp_floor may be better if fitting low-SNR tail of data is desired.
- Often you don't need to fit an entire ringdown, many cycles with high SNR is usually sufficient.

## Keyword arguments for "fit_single_ringdown":
- f_fit_iter (_default val 0_): Number of times to perform software lockin, phase fit, and update the frequency before moving on to amplitude fit.
- method (_default 'logfit'_): Method for fitting the amplitude
- decimation (_default 50_): how much to downsample when performing software lockin
- harmonics (_default 10_): How many harmonics to calculate during software lockin, used to make the THD calculation
- startbuff (_default 0.01_): what fraction of data to throw out at the start of the timestream
- endbuff (_default 0.01_): what fraction of data to throw out at the end of the timestream (trumped by cutoffs from detuning and THD)
- minPhiSamples (_default 10_): minimum number of phase data points to use in the fit
- max_expected_detuning (_default 20000_): maximum jump between points allowed in phase fit
- max_THD (_default 10_): maximum value of total harmonic distortion allowed in amplitude fit. THD is defined as $\frac{\sqrt{\sum_n (r^2(nf_0))}}{r(f_0)}$ where $r(nf_0)$ is the amplitude at the $n$th harmonic of the lockin frequency. It's a bit of a hack to use this as a proxy for low SNR data cutoff... but it works
- start_max_THD (_defaults to max_THD_): separate cutoff used for the start of amplitude data in the fit. Useful to set tight value for this (~0.5 or below) to eliminate transient higher harmonics at the start of a ringdown due to transient effects.
- printThings (_default TRUE_): allow print outputs
- showPLTs (_default TRUE_): allow plots to be shown

## Let's generate some example data

To demonstrate the fitting code, let's make some fake ringdown data by multiplying a sine wave by an exponential decay with time constant $\tau$. Here, I've arbitrarily picked a frequency of 670.123 kHz and a Q of 136. I've also added some gaussian noise.

```python
f = 670.123e3
Q = 136
tau = Q/(f*np.pi)
samplerate = 50e6
ts = 2e-4

t = np.linspace(0, ts, int(samplerate*ts))
amp = np.sin(2*np.pi*f*t+np.pi/17)*(np.exp(-1*t/tau))+0.05*np.random.randn(len(t))
```

```python
fig, ax = mpl.subplots(figsize=(10,5))
ax.plot(t*1e3, amp)
ax.set_ylabel('Signal')
ax.set_xlabel('time (ms)')
ax.set_title('Example ringdown data')
fig.tight_layout()
```

```python
importlib.reload(rd)
Q_fit, F_fit = rd.fit_single_ringdown(t, amp, 670e3, decimation=20, method='logfit', startbuff = 0, endbuff = 0)
print(f'Fitted values: Q = {Q_fit}, F = {F_fit}')
print(f'Q fitted with {100*np.abs(Q_fit-Q)/Q:.2f}% error')
print(f'Freq fitted with {100*np.abs(F_fit-f)/f:.10f}% error')
```

## Ok-- first results.

The fit does pretty well! Now, some caveats-- we've only sampled when there is relatively good SNR, before the signal disappears into the noise floor. Let's try with a "longer" timestream, and see if we can use more of the low SNR data. for now, we'll set the amplitude fit cutoff unreasonably high to try and use even the "bad" data.

```python
f = 670.123e3
Q = 136
tau = Q/(f*np.pi)
samplerate = 50e6
ts = 5e-4

t = np.linspace(0, ts, int(samplerate*ts))
amp = 2*np.sin(2*np.pi*f*t+np.pi/17)*(np.exp(-1*t/tau))+0.2*np.random.randn(len(t))
```

```python
fig, ax = mpl.subplots()
ax.plot(t*1e3, amp)
ax.set_ylabel('Signal')
ax.set_xlabel('time (ms)')
ax.set_title('Example ringdown data')
fig.tight_layout()
```

```python
Q_fit, F_fit = rd.fit_single_ringdown(t, amp, 670e3, decimation=20, method='logfit', startbuff = 0, endbuff = 0, max_THD = 100)
print(f'Fitted values: Q = {Q_fit}, F = {F_fit}')
print(f'Q fitted with {100*np.abs(Q_fit-Q)/Q:.2f}% error')
print(f'Freq fitted with {100*np.abs(F_fit-f)/f:.10f}% error')
```

## But!

There's another option-- switch the fitting method to "exp_floor". This is a slightly more computationally intensive fit, because we are fitting to a more complicated exponential decay into a floor. Let's still leave the THD cutoff out of it and try to fit the whole timestream.

```python
importlib.reload(rd)
Q_fit, F_fit = rd.fit_single_ringdown(t, amp, 670e3, decimation=20, method='exp_floor', startbuff = 0, endbuff = 0, max_expected_detuning = 50000, max_THD=100)
print(f'Fitted values: Q = {Q_fit}, F = {F_fit}')
print(f'Q fitted with {100*np.abs(Q_fit-Q)/Q:.2f}% error')
print(f'Freq fitted with {100*np.abs(F_fit-f)/f:.10f}% error')
```

## Ok! That's better! 6% error is...ok, and we can do even better by setting appropriate THD cutoffs.

```python
importlib.reload(rd)
Q_fit, F_fit = rd.fit_single_ringdown(t, amp, 670e3, decimation=20, method='exp_floor', startbuff = 0, endbuff = 0, max_expected_detuning = 50000, max_THD=2)
print(f'Fitted values: Q = {Q_fit}, F = {F_fit}')
print(f'Q fitted with {100*np.abs(Q_fit-Q)/Q:.2f}% error')
print(f'Freq fitted with {100*np.abs(F_fit-f)/f:.10f}% error')
```

 *if we're lucky...below, I show what happens if we generate data with the same noise params (but different random noise) and try the fit over and over. 

```python
f = 670.123e3
Q = 50
tau = Q/(f*np.pi)
samplerate = 50e6
ts = 1e-4

t = np.linspace(0, ts, int(samplerate*ts))
amp = 2*np.sin(2*np.pi*f*t+np.pi/17)*(np.exp(-1*t/tau))+0.5*np.random.randn(len(t))

fig, ax = mpl.subplots(figsize=(10,5))
ax.plot(t*1e3, amp)
ax.set_ylabel('Signal')
ax.set_xlabel('time (ms)')
ax.set_title('Example ringdown data')
fig.tight_layout()
```

```python
importlib.reload(rd)
Q_fit, F_fit = rd.fit_single_ringdown(t, amp, 670e3, f_fit_iter = 2, decimation=50, method='exp_floor', startbuff = 0, endbuff = 0, max_expected_detuning = 100000, max_THD=5)
print(f'Fitted values: Q = {Q_fit}, F = {F_fit}')
print(f'Q fitted with {100*np.abs(Q_fit-Q)/Q:.2f}% error')
print(f'Freq fitted with {100*np.abs(F_fit-f)/f:.10f}% error')
```

Is this result consistent? Let's try to fit a bunch (500!) of data sets with the same params.

```python
Qlist_log = []
Flist_log = []
Qlist_floor = []
Flist_floor = []
num = 500
for i in np.arange(num):
    amp = 2*np.sin(2*np.pi*f*t+np.pi/17)*(np.exp(-1*t/tau))+0.5*np.random.randn(len(t))
    Q_fit, F_fit = rd.fit_single_ringdown(t, amp, 670e3, f_fit_iter = 1, decimation=20, method='logfit', startbuff = 0, endbuff = 0, max_expected_detuning = 100000, max_THD=3, printThings=False, showPLTs=False)
    Qlist_log.append(Q_fit)
    Flist_log.append(F_fit)
    Q_fit, F_fit = rd.fit_single_ringdown(t, amp, 670e3, f_fit_iter = 1, decimation=20, method='exp_floor', startbuff = 0, endbuff = 0, max_expected_detuning = 100000, max_THD=3, printThings=False, showPLTs=False)
    Qlist_floor.append(Q_fit)
    Flist_floor.append(F_fit)
```

```python
mpl.close('all')
fig, ax = mpl.subplots(2,1, figsize=(10,5))
ax[0].scatter(np.arange(num), Qlist_log, c='red', label='Log fit')
ax[0].scatter(np.arange(num), Qlist_floor, c = 'blue', label='Exp floor fit')
ax[0].hlines(50, 0, num, color='black', label='Actual Q')
ax[0].hlines(np.mean(Qlist_log), 0, num, color='red')
ax[0].hlines(np.mean(Qlist_floor), 0, num, color='blue')
ax[1].scatter(np.arange(num), Flist_log, c='red', label='Log fit')
ax[1].scatter(np.arange(num), Flist_floor, c='blue', label='Exp floor fit')
ax[1].hlines(f, 0, num, color='black', label='Actual F')
ax[1].hlines(np.mean(Flist_log), 0, num, color='red')
ax[1].hlines(np.mean(Flist_floor), 0, num, color='blue')
ax[0].legend()
fig.tight_layout()
print((np.mean(Qlist_log)-50)/50)
print((np.mean(Qlist_floor)-50)/50)
```

```python
mpl.close('all')
fig, ax = mpl.subplots(1, 2, figsize=(10,5))
ax[0].hist(Qlist_log, alpha=0.5, color='red',label='Log fit')
ax[0].hist(Qlist_floor, alpha=0.5, color='blue', label='Exp floor fit')
ax[0].vlines(50, 0, 100, color='black', label='Actual Q')
ax[0].legend()
ax[1].hist(Flist_log, alpha=0.5, color='red',label='Log fit')
ax[1].hist(Flist_floor, alpha=0.5, color='blue',label='Exp floor fit')
ax[1].vlines(f, 0, 100, color='black', label='Actual F')
ax[1].legend()
```

## Takeaways from noisy data:
Logfit tends to overestimate the Q, while exp_floor underestimates the Q with a left tail on the distribution.

## Now trying to deal with saturation effects:

Sometimes at the start of a ringdown timestream, there will be significant amplitude in the second/higher harmonics. This can be caused by saturation effects, such as the SQUID response being driven nonlinear. To account for this, there is a separate THD cutoff at the *beginning* of the amplitude fit. This delays the start of the fit, giving transient higher harmonics time to decay away.

```python
f = 670.123e3
Q1 = 50
Q2 = 8
tau = Q1/(f*np.pi)
tau2 = Q2/(2*f*np.pi)
samplerate = 50e6
ts = 1e-4

t = np.linspace(0, ts, int(samplerate*ts))
amp1 = np.sin(2*np.pi*f*t)*(np.exp(-1*t/tau))
amp2 = 2*np.sin(4*np.pi*f*t)*(np.exp(-1*t/tau2))+2*np.sin(6*np.pi*f*t)*(np.exp(-1*t/tau2))
amp = amp1+amp2 + 0.1*np.random.randn(len(t))

fig, ax = mpl.subplots(figsize=(10,5))
ax.plot(t*1e3, amp1+5, label= r'component at $f_0$')
ax.plot(t*1e3, amp2, label=r'component at $2f_0$')
ax.plot(t*1e3, amp-5, label='combined repsonse, with added noise')
ax.set_yticks([])
ax.set_xlabel('time (ms)')
ax.set_title('Example ringdown data')
ax.legend(loc='center right')
fig.tight_layout()
```

```python
importlib.reload(rd)
Q_fit, F_fit = rd.fit_single_ringdown(t, amp, 670e3, f_fit_iter = 1, decimation=10, method='logfit', startbuff = 0, endbuff = 0, max_expected_detuning = 100000, start_max_THD=0.5, max_THD=1)
print(f'Fitted values: Q = {Q_fit}, F = {F_fit}')
print(f'Q fitted with {100*np.abs(Q_fit-Q1)/Q1:.2f}% error')
print(f'Freq fitted with {100*np.abs(F_fit-f)/f:.10f}% error')
```
